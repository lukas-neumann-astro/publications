# Packages

```python
# Matplotlib setup
from matplotlib_setup import mpl_setup

# Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as font_manager
import matplotlib.gridspec as gridspec
import seaborn as sns
```

# Ancillary functions

```python
# lighten/darken colour
def lighten_color(color, amount=0.5):
    """
    Lightens the given color by multiplying (1-luminosity) by the given amount.
    Input can be matplotlib color string, hex string, or RGB tuple.

    Examples:
    >> lighten_color('g', 0.3)
    >> lighten_color('#F034A3', 0.6)
    >> lighten_color((.3,.55,.1), 0.5)
    """
    import matplotlib.colors as mc
    import colorsys
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

# Setup

```python
# directories
wd = '/Users/lneumann/Documents/'
database_dir = wd + 'Products/dense_gas_letter/database/'
tab_dir =  wd + 'Products/dense_gas_letter/tables/'
plots_dir = wd + 'Products/dense_gas_letter/plots/main/'

# load dense gas literature
fname_table = database_dir + 'gao_solomon_literature_compilation_2024.csv'
table = pd.read_csv(fname_table)

# conversions
tir2sfr = 1.48e-10  # [Msun/yr] Murphy+11
alpha_hcn = 15  # [M_Sun/pc^2/(K km/s)] (Schinnerer & Leroy 2024)

# SFE median across literature compilation (2024)
fname_gs_table = tab_dir + 'GS_relation_per_regime.csv'
gs_table = pd.read_csv(fname_gs_table)
sfe_median = np.array(gs_table[gs_table['data']=='combined']['SFR/HCN (median)'])[0]
sfe_scatter = np.array(gs_table[gs_table['data']=='combined']['scatter'])[0]

# ALMOND + EMPIRE
marker_almond = 'h'
marker_empire = 'o'
color_almond = 'tab:blue'
color_empire = 'tab:red'
refs_dense = ['Jiménez-Donaire19', 'Neumann23']
marker_dense = [marker_empire, marker_almond]
color_dense = [color_empire, color_almond]
labels_dense = ['Jiménez-Donaire et al. 2019 (EMPIRE)', 'Neumann et al. 2023b (ALMOND)']

# sub-samples
table_clouds = table[table['Type'] == 'clouds and clumps']
table_resolved = table[table['Type'] == 'parts of galaxies']
table_galaxies = table[table['Type'] == 'entire galaxies']
color_clouds = 'tab:green'
color_resolved = 'tab:olive'
color_galaxies = 'tab:pink'
color_mw_cmz = 'orange'

# references
refs_clouds = ['Chin97','Chin98','Brouillet05','Wu10','Lada12','Buchbender13','Evans14','Stephens16','Braine17']
colors_clouds = ['xkcd:spring green', 'xkcd:pale yellow', 'xkcd:pale pink', 'xkcd:powder blue', 'xkcd:greyish green', 'xkcd:pumpkin', 'xkcd:light red', 'xkcd:ochre', 'xkcd:avocado']
markers_clouds = ['s'] * len(refs_clouds)

refs_resolved = ['Kepley14', 'Usero15', 'Bigiel15', 'Chen17', 'Gallagher18', 'Jiménez-Donaire19', 'Querejeta19', 'Bešlić21', 'Eibensteiner22', 'Sánchez-García22', 'Neumann23', 'Stuber23', 'Bešlić24', 'Neumann24']
colors_resolved = ['tab:green', 'tab:purple', 'tab:brown', 'limegreen', 'tab:pink', color_empire, 'tab:orange', 'tab:gray', 'tab:cyan', 'tan', color_almond, 'tab:olive', 'xkcd:dark pink', 'm']
markers_resolved = ['o'] * len(refs_resolved)

refs_galaxies = ['Gao04','Gao07','Krips08','Gracia-Carpio08','Juneau09','Crocker12','Garcia-Burillo12','Privon15','Rybak22']
colors_galaxies = ['xkcd:sky blue', 'xkcd:teal', 'xkcd:burnt orange', 'xkcd:mauve', 'xkcd:mustard', 'xkcd:coral', 'xkcd:light green', 'xkcd:dull blue', 'xkcd:aubergine']
markers_galaxies = ['D'] * len(refs_galaxies)

# combine refs
refs_tot = refs_clouds + refs_resolved + refs_galaxies
colors_tot = colors_clouds + colors_resolved + colors_galaxies
markers_tot = markers_clouds + markers_resolved + markers_galaxies

# sort by publication date
refs_tot_yr = [np.array(table['Publication date'][table['Name']==ref])[0] for ref in refs_tot]
refs_tot_yr = [float(ref.split('/')[1]) + float(ref.split('/')[0])/12 for ref in refs_tot_yr]
refs_tot_yr, refs_tot, colors_tot, markers_tot = zip(*sorted(zip(refs_tot_yr, refs_tot, colors_tot, markers_tot)))

# secondary y-axis for SFR to TIR
def sfr_to_tir(sfr):
    return sfr / tir2sfr
def tir_to_sfr(tir):
    return tir * tir2sfr

# secondary x-axis for HCN to Mdense
def hcn_to_mdense(hcn):
    return hcn * alpha_hcn
def mdense_to_hcn(mdense):
    return mdense / alpha_hcn
```

# Plotting

```python
# toogle save figure
savepng = False
savepdf = True

# matplotlib setup
mpl_setup(figtype='paper-1/1', ms=4)

# create figure
fig = plt.figure()
gs = gridspec.GridSpec(2, 1, figure=fig, wspace=0, hspace=0, height_ratios=[2,1])
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

# handles for legend
handles_clouds = []
handles_resolved = []
handles_galaxies = []
handles_clouds_grey = []
handles_resolved_grey = []
handles_galaxies_grey = []
handles_dense = []

# iterate over references
for ref, color, marker in zip(refs_tot, colors_tot, markers_tot):

    # get data
    ids = (table['Name']==ref)
    x = table['L_HCN [K.km/s.pc2]'][ids]
    y = table['SFR [Msun/yr]'][ids]

    # plot data
    label = ref[:-2] + ' et al. 20' + ref[-2:] if int(ref[-2:]) < 50 else ref[:-2] + ' et al. 19' + ref[-2:]
    if (ref=='Neumann23'):
        label = 'Neumann et al. 2023b'
    handle, = ax1.plot(x, y, marker=marker, mfc=color, mec='k', mew=0.5, ls='none', label=label, zorder=3)
    if (ref=='Jiménez-Donaire19'):
        handle, = ax2.plot(x, y/x, marker=marker, mfc=color, mec='k', mew=0.5, ls='none', label=label, zorder=12)
        handles_dense.append(handle)
    if (ref=='Neumann23'):
        handle, = ax2.plot(x, y/x, marker=marker, mfc=color, mec='k', mew=0.5, ls='none', label=label, zorder=11)
        handles_dense.append(handle)
    else:
        handle_grey, = ax2.plot(x, y/x, marker=marker, mfc=lighten_color('grey', amount=0.5), mec='grey', mew=0.5, ls='none', label=label, zorder=3)
        
    # append handle to sample list
    if marker == 's':
        handles_clouds.append(handle)
        handles_clouds_grey.append(handle_grey)
    elif marker == 'o':
        handles_resolved.append(handle)
        handles_resolved_grey.append(handle_grey)
    elif marker == 'D':
        handles_galaxies.append(handle)
        handles_galaxies_grey.append(handle_grey)

################

## PLOT MW CMZ
x = table['L_HCN [K.km/s.pc2]'][table['Name']=='Henshaw23']
y = table['SFR [Msun/yr]'][table['Name']=='Henshaw23']
handle_cmz, = ax1.plot(x, y, marker='*', ms=10, mfc=color_mw_cmz, mec='k', mew=0.5, ls='none', label='Jones et al. 2012 &\n Barnes et al. 2017', zorder=5)
handle_cmz_grey, = ax2.plot(x, y/x, marker='*', ms=10, mfc=lighten_color('grey', amount=0.5), mec='grey', mew=0.5, ls='none', label='Jones et al. 2012 &\n Barnes et al. 2017', zorder=5)


################

# plot median SFE line
x_values = np.logspace(-2,12,100)
# SFR vs HCN
y_values = x_values*10**sfe_median 
ax1.plot(x_values, y_values, c='k')
ax1.plot(x_values, y_values*10**sfe_scatter, c='k', ls='dashed')
ax1.plot(x_values, y_values/10**sfe_scatter, c='k', ls='dashed')
ax1.fill_between(x_values, y_values*10**sfe_scatter, y_values/10**sfe_scatter, color='grey', alpha=0.3, lw=0)
# SFR/HCN vs HCN
y_values = np.full_like(x_values, 10**sfe_median)
ax2.plot(x_values, y_values, c='k')
ax2.plot(x_values, y_values*10**sfe_scatter, c='k', ls='dashed')
ax2.plot(x_values, y_values/10**sfe_scatter, c='k', ls='dashed')
ax2.fill_between(x_values, y_values*10**sfe_scatter, y_values/10**sfe_scatter, color='grey', alpha=0.3, lw=0)


################ CONTOURS

# SUB-SAMPLES
for x, y, c in zip([table_clouds['L_HCN [K.km/s.pc2]'], table_resolved['L_HCN [K.km/s.pc2]'], table_galaxies['L_HCN [K.km/s.pc2]']],
                [table_clouds['SFR [Msun/yr]'], table_resolved['SFR [Msun/yr]'], table_galaxies['SFR [Msun/yr]']],
                [color_clouds, color_resolved, color_galaxies]):
    mask = (y > 1e-10) & (x>1e-5)  # exclude outlier
    sns.kdeplot(x=x[mask], y=(y/x)[mask], log_scale=True, levels=[0.1], color='grey', ax=ax2, zorder=20)

# EMPIRE & ALMOND
ids = (table_resolved['Name']=='Jiménez-Donaire19')
x = table_resolved['L_HCN [K.km/s.pc2]'][ids]
y = table_resolved['SFR [Msun/yr]'][ids]
sns.kdeplot(x=x, y=y/x, log_scale=True, levels=[0.1], color=color_empire, ax=ax2, zorder=22)

ids = (table_resolved['Name']=='Neumann23')
x = table_resolved['L_HCN [K.km/s.pc2]'][ids]
y = table_resolved['SFR [Msun/yr]'][ids]
sns.kdeplot(x=x, y=y/x, log_scale=True, levels=[0.1], color=color_almond, ax=ax2, zorder=21)




################

# secondary x-axis for HCN-Mdense
secxax = ax1.secondary_xaxis('top', functions=(hcn_to_mdense, mdense_to_hcn)) 
secxax.set_xlabel('$M_\mathrm{dense}$ [$\mathrm{M}_{\odot}$]', labelpad=8)
ax1.tick_params(which='both', top=False)

# secondary y-axis for SFR-IR
secyax1 = ax1.secondary_yaxis('right', functions=(sfr_to_tir, tir_to_sfr)) 
secyax1.set_ylabel('$L_\mathrm{IR}^\mathrm{equiv}$ [L$_{\odot}$]')

# secondary y-axis for SFR/HCN-IR/HCN
secyax2 = ax2.secondary_yaxis('right', functions=(sfr_to_tir, tir_to_sfr)) 
secyax2.set_ylabel(r'$\frac{L_\mathrm{IR}^\mathrm{equiv}\;[L_{\odot}]}{L_{\mathrm{HCN}(1-0)}\;[\mathrm{K}\,\mathrm{km}\,\mathrm{s}^{-1}\,\mathrm{pc}^2]}$', size='x-large')

# axis labels
ax2.set_xlabel('$L_{\mathrm{HCN}(1-0)}$ [K$\,$km$\,\mathrm{s}^{-1}\,\mathrm{pc}^2$]')
ax1.set_ylabel('SFR [$\mathrm{M}_{\odot}\,\mathrm{yr}^{-1}$]')
ax2.set_ylabel(r'$\frac{\mathrm{SFR}\;[\mathrm{M}_{\odot}\,\mathrm{yr}^{-1}]}{L_{\mathrm{HCN}(1-0)}\;[\mathrm{K}\,\mathrm{km}\,\mathrm{s}^{-1}\,\mathrm{pc}^2]}$', size='x-large')

# general settings
for ax in [ax1, ax2]:
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(1e-1, 3e11)
    ax.set_xticks(np.logspace(0,10,6))
    ax.tick_params(which='both', right=False)
ax1.set_ylim(1e-1*10**sfe_median, 3e11*10**sfe_median)
ax2.set_ylim(5e-9, 2e-6)
ax1.tick_params(labelbottom=False)
# ax1.grid()



################

# legend (ax1)
mpl.rcParams['text.usetex'] = False
font = font_manager.FontProperties(family='Arial', style='normal', size=6)
title_font = font_manager.FontProperties(family='Arial', weight='bold', style='normal', size=10)
legend1 = ax1.legend(title='Clouds & Clumps', title_fontproperties=title_font, handles=handles_clouds, loc='lower right', 
                     prop=font, labelspacing=0.1, handlelength=1,
                     # edgecolor=color_clouds, 
                     bbox_to_anchor=[0.78, 0]
                    )
ax1.add_artist(legend1)

legend2 = ax1.legend(title='Parts of Galaxies', title_fontproperties=title_font, handles=handles_resolved, loc='upper left', 
                     prop=font, labelspacing=0.1, handlelength=1,
                     # edgecolor=color_resolved, 
                    )
ax1.add_artist(legend2)

legend3 = ax1.legend(title='Entire Galaxies', title_fontproperties=title_font, handles=handles_galaxies, loc='lower right', 
                     prop=font, labelspacing=0.1, handlelength=1,
                     # edgecolor=color_galaxies, 
                    )
ax1.add_artist(legend3)

legend4 = ax1.legend(title='MW CMZ', title_fontproperties=title_font, handles=[handle_cmz], loc='upper left', 
                     prop=font,labelspacing=0.3, handletextpad=1, handlelength=1.5,
                     bbox_to_anchor=[0.24, 1])
# legend (ax2)
legend5 = ax2.legend(prop=font,
                    # title='Literature', title_fontproperties=title_font, 
                   handles = [handles_clouds_grey[0], handles_resolved_grey[0], handles_galaxies_grey[0], handle_cmz_grey], 
                   labels = ['Clouds & Clumps', 'Parts of Galaxies', 'Entire Galaxies', 'MW CMZ'],
                   labelspacing=0.1, handlelength=1, ncol=2,
                   loc='lower right')
legend5.set_zorder(50)
ax2.add_artist(legend5)
legend6 = ax2.legend(prop=font, handles = handles_dense, labels = labels_dense,
                   labelspacing=0.1, handlelength=1,
                   loc='upper right')
legend6.set_zorder(50)
mpl.rcParams['text.usetex'] = True 


#####################################

# save plot
savepath = plots_dir + 'GS_relation_literature_compilation'
if savepng:
    plt.savefig(savepath + '.png')
if savepdf:
    plt.savefig(savepath + '.pdf')

plt.show()
```
